```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
```

```python
df = pd.read_csv('data/all/music_data.csv')
df_meta = pd.read_csv('data/all/metadata.csv', parse_dates=['album_date_released'])

df = df.drop(columns=['track_id'])
df['popularity'] = df_meta['track_listens']
df['release'] = df_meta['album_date_released']

df
```

#### Splits

```python
train, test = train_test_split(df, test_size=0.2, random_state=1)

train.to_csv('data/split/train.csv', index=False)
test.to_csv('data/split/test.csv', index=False)
```

#### Folds

```python
train = train.reset_index(drop=True)

kf = KFold(n_splits=5)
i = 1
for train_index, test_index in kf.split(train):
    print("TRAIN:", train_index, "TEST:", test_index)
    f_train, f_test = train.iloc[train_index], train.iloc[test_index]
    f_train.to_csv(f'data/fold/f{i}_train.csv', index=False)
    f_test.to_csv(f'data/fold/f{i}_test.csv', index=False)
    i+=1
```
